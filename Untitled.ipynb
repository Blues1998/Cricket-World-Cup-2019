```python
#Match 14
#India vs Australia match held on 9th June 2019
stats = pd.read_excel('Match 14/Stats per over.xlsx')

over = list(stats['Over'])
runs_Ind = list(stats['Runs Ind'])
runs_Aus = list(stats['Runs Aus'])
wickets_Ind = list(stats['Wicket at Run IN'])
wickets_Aus = list(stats['Wicket at Run AUS'])
fig = plt.figure()
fig.set_size_inches(10, 4)
plt.figure(figsize=(5,5))
# fig, ax = plt.subplots()
ax = fig.add_axes([0,0,1,1])
ax.plot(over,runs_Ind,'#03A9F4',linestyle='-',label='India')
ax.plot(over,runs_Aus,'#FFEB3B',linestyle='-',label='Australia')
ax.plot(over,wickets_Ind,'#03A9F4',marker='o',linewidth='4',markersize=11,label='IN Wickets')
ax.plot(over,wickets_Aus,'#CDDC39',marker='o',linewidth='4',markersize=11,label='AU Wickets')
ax.set_title('Australia Run Chase')
ax.set_ylabel('Runs')
ax.set_xlabel('Overs')
# plt.grid(True,color='blue',axis='both',linestyle=':')
ax.legend(loc=5)
```

```python
import tkinter as tk
HEIGHT=600
WIDTH=1000

root=tk.Tk()
root.state('zoomed')
root.resizable(0,0)
#Background Canvas and Frame

canvas=tk.Canvas(root,height=HEIGHT,width=WIDTH)
canvas.pack()
frame=tk.Frame(root,bg='#009688').place(relx=0,rely=0,relheight=1,relwidth=1)
head=tk.Label(frame,bg='#ffb84d',text='CWC 2k19',font='Times 15 bold',bd=5).place(relx=0,rely=0,relheight=.1,relwidth=1)

    

'''Functions for frames and windows'''

#Function to open Window for Indian team
def wd_Ind():
    def m8():
        im_m8 = ImageTk.PhotoImage(Image.open("Match 8\IMGS\chase.png"))
        lbl_m8 = tk.Label(wd_India, image = im_m8).place(relx=.1,rely=.35)
        lbl_m8.image = im_m8
    wd_India = tk.Toplevel(root,height=HEIGHT,width=WIDTH,bg='#03A9F4')
    wd_India.title('India')
    wd_India.state('zoomed')
    wd_India.resizable(0,0)
    def m14():
        im_m14 = ImageTk.PhotoImage(Image.open("Match 14\IMGS\chase.png"))
        lbl_m14 = tk.Label(wd_India, image = im_m14).place(relx=.1,rely=.35)
        lbl_m14.image = im_m14
    btn_match8 = tk.Button(wd_India,text='Match 8\nIndia vs South Africa',bg='#CDDC39',font='Times 12 bold',bd=0.5,command=m8).place(relx=.05,rely=.1,relheight=.08,relwidth=.20)
    btn_match14 = tk.Button(wd_India,text='Match 14\nIndia vs Australia',bg='#FFEB3B',font='Times 12 bold',bd=0.5,command=m14).place(relx=.27,rely=.1,relheight=.08,relwidth=.20)
    
    
#Function to open Window for Australian team
def wd_Aus():
    #Function for Match 14
    def m14():
        im_m14 = ImageTk.PhotoImage(Image.open("Match 14\IMGS\chase.png"))
        lbl_m14 = tk.Label(wd_Australia, image = im_m14).place(relx=.1,rely=.35)
        lbl_m14.image = im_m14
        
    #Opening the Window for Australia team analysis
    wd_Australia = tk.Toplevel(root,height=HEIGHT,width=WIDTH,bg='#FFEB3B')
    
    #Setting title for title bar
    wd_Australia.title('Australia')
    
    #Opening Window in maximised view
    wd_Australia.state('zoomed')
    
    #Disabling the maximise button
    wd_Australia.resizable(0,0)
    btn_match14 = tk.Button(wd_Australia,text='Match 14\nIndia vs Australia',bg='#03A9F4',font='Times 12 bold',bd=0.5,command=m14).place(relx=.05,rely=.1,relheight=.08,relwidth=.20)
    
    
def wd_NZe():
    wd_NewZ = tk.Toplevel(root,height=HEIGHT,width=WIDTH,bg='#212121')
    wd_NewZ.title('New Zealand')
    
    
def wd_SAf():
    def m8():
        im_m8 = ImageTk.PhotoImage(Image.open("Match 8\IMGS\chase.png"))
        lbl_m8 = tk.Label(wd_SAfrica, image = im_m8).place(relx=.1,rely=.35)
        lbl_m8.image = im_m8
    wd_SAfrica = tk.Toplevel(root,height=HEIGHT,width=WIDTH,bg='#CDDC39')
    
    #Setting title for title bar
    wd_SAfrica.title('South Africa')
    
    #Opening Window in maximised view
    wd_SAfrica.state('zoomed')
    
    #Disabling the maximise button
    wd_SAfrica.resizable(0,0)
    btn_match8 = tk.Button(wd_SAfrica,text='Match 8\nIndia vs South Africa',bg='#03A9F4',font='Times 12 bold',bd=0.5,command=m8).place(relx=.05,rely=.1,relheight=.08,relwidth=.20)
    
    
def wd_Ban():
    wd_Bangladesh = tk.Toplevel(root,height=HEIGHT,width=WIDTH,bg='#4CAF50')
    wd_Bangladesh.title('Bangladesh')
    

#Team Buttons

btn_Aus = tk.Button(frame,text='Australia',bg='#FFEB3B',font='Times 15 bold',bd=0.5,command=wd_Aus).place(relx=.2,rely=.3,relheight=.08,relwidth=.28)
btn_Ind = tk.Button(frame,text='India',bg='#03A9F4',font='Times 15 bold',bd=0.5,fg='White',command=wd_Ind).place(relx=.2,rely=.4,relheight=.08,relwidth=.28)
btn_NZe = tk.Button(frame,text='New Zealand',bg='#212121',font='Times 15 bold',bd=0.5,fg='White',command=wd_NZe).place(relx=.2,rely=.6,relheight=.08,relwidth=.28)
btn_SAf = tk.Button(frame,text='South Africa',bg='#CDDC39',font='Times 15 bold',bd=0.5,command=wd_SAf).place(relx=.2,rely=.5,relheight=.08,relwidth=.28)
btn_Ban = tk.Button(frame,text='Bangladesh',bg='#4CAF50',font='Times 15 bold',bd=0.5,command=wd_Ban).place(relx=.2,rely=.7,relheight=.08,relwidth=.28)
btn_Pak = tk.Button(frame,text='Pakistan',bg='#2E7D32',font='Times 15 bold',bd=0.5,fg='White').place(relx=.5,rely=.3,relheight=.08,relwidth=.28)
btn_WIn = tk.Button(frame,text='West Indies',bg='#C62828',font='Times 15 bold',bd=0.5).place(relx=.5,rely=.4,relheight=.08,relwidth=.28)
btn_Eng = tk.Button(frame,text='England',bg='#283593',font='Times 15 bold',bd=0.5,fg='White').place(relx=.5,rely=.5,relheight=.08,relwidth=.28)
btn_Afg = tk.Button(frame,text='Afghanistan',bg='#2196F3',font='Times 15 bold',bd=0.5).place(relx=.5,rely=.6,relheight=.08,relwidth=.28)
btn_SLa = tk.Button(frame,text='Sri Lanka',bg='#0288D1',font='Times 15 bold',bd=0.5,fg='White').place(relx=.5,rely=.7,relheight=.08,relwidth=.28)


root.mainloop()
```
